```python
from Border import *
#from BuildingBlock import *
#from Coupler import *
from Modulator import *
from LayerCSEM import *
from EdgeCouplerDoubleTaper import *
from nazca.interconnects import Interconnect
import nazca as nd
# Layers as defined by CSEM
nd.add_layer(name='WG1', layer=11)


nd.add_layer(name='WG2_Opening', layer=25)
nd.add_layer(name='WG2', layer=21) #WG2_SE
nd.add_layer(name='MET1', layer=51)
nd.add_layer(name='CLD_OPEN', layer=60)
nd.add_layer(name='DEEP_GRID', layer=70)
nd.add_layer(name='TXT_LOGO', layer=22) # "WG2_SE_TXT"
nd.add_layer(name='EFF AREA', layer=2)
nd.add_layer(name='PAYLOAD', layer=0)

# Create a CROSS-SECTIONS for Passive (XS1) and Active (XS2)
nd.add_xsection(name='XS1')
nd.add_xsection(name='XS2')

# WAVEGUIDE PARAMETER RULES
width_WG = 0.8
rad_bend = 20

# Define the cross section to use for the PASSIVE WG
# Define PASSIVE interconnect according to waveguide parameter rules
nd.add_layer2xsection(xsection='XS1', layer='WG1')
nd.add_layer2xsection(xsection='XS2', layer='WG1')
nd.add_layer2xsection(xsection='XS2', layer='MET1')
passive = Interconnect(xs='XS1', width=width_WG, radius=rad_bend)
active = Interconnect(xs='XS2', width=width_WG, radius=rad_bend)





def myMMI(length=400, posX = 100 , posY = -700):
    with nd.Cell(name=f'myMMI_{length}') as myMMI:
#         myMMI.autobbox = False
#         parametersDCoupler = {"waveguide_width": 2.0, "length": 32.704, "radius": 15, "spacing": 12, "gap": 0.8, "arrow": False}
#         directionCoupler1 = DirectionalCoupler(**parametersDCoupler).put(-100, -20)
#         waveguide1 = waveguide(length = 180).put(posX+70, posY+ 0)
#         directionCoupler2 = DirectionalCoupler(**parametersDCoupler).put(posX+420, posY-20, flop=True)
#         passive.sbend_p2p(pin1=directionCoupler1.pin['b1'], pin2=waveguide1.pin['a0']).put()
#         passive.sbend_p2p(pin1=directionCoupler1.pin['b0'], pin2=waveguide1.pin['a1']).put()
#         passive.sbend_p2p(pin1=waveguide1.pin['b0'], pin2=directionCoupler2.pin['b1']).put()
#         passive.sbend_p2p(pin1=waveguide1.pin['b1'], pin2=directionCoupler2.pin['b0']).put()
        parameters = {"width": 0.8, "signal_length": 1000, "signal_gap": 14.6, "signal_width": 4.5, "arrow": True}
    
        
        m = Modulator(**parameters).put(posX, posY)
        nd.Pin('a0', pin=m.pin['c0']).put()
        nd.Pin('a1', pin=m.pin['c1']).put()
        nd.Pin('b0', pin=m.pin['d0']).put()
        nd.Pin('b1', pin=m.pin['d1']).put()
    return myMMI

def main():
    parametersBorder = {"length": 10000, "width": 5000}
    parametersEdge = {"width": 0.8, "width1": 0.280, "width2": 1.2, "width3": 3, "width4": 10, "width_top_taper": 0.05, "length1": 100, "length2": 100, "length3": 10, "arrow": False}
    parametersEdgeEntry = {"width": 0.8, "width1": 0.280, "width2": 1.2, "width3": 3, "width4": 10, "width_top_taper": 0.05, "length1": 28, "length2": 100, "length3": 10, "arrow": False}
    
    Border(**parametersBorder).put(5000, -2500)
    nd.nazca_logo(layers={'ring':1, 'box':20, 'bird':11}).put(300, -300)

    nd.image('ucl-logo-black-on-grey.jpg', threshold=0.5, size=200, pixelsize=5).put(1400, -300, flop = False)

    #nd.image('UCL_logo.png', threshold=0.6).put(-700, -300)
    message1 = 'V.Liutarevich'
    nd.text(text=message1, height=90, layer=22, align='cc').put(3000, -300)
    passive.strt(length=9928, xs='XS1').put(0, -550)
    EdgeCouplerDoubleTaper(**parametersEdge).put()
    
    
    EdgeCouplerDoubleTaper(**parametersEdgeEntry).put(0, -550, flop=True)
    my_mmi1 = myMMI(posX = 0 , posY = -700).put(0, -700)
    finalConnect1 = passive.sbend().put(my_mmi1.pin['b0'])
    passive.strt(length=8383.35, xs='XS1').put(finalConnect1.pin['b0'])
    EdgeCouplerDoubleTaper(**parametersEdge).put()
    message2 = 'UCL Photonics Group'
    nd.text(text=message2, height=90, layer=22, align='cc').put(4000, -300)
    
    EdgeCouplerDoubleTaper(**parametersEdgeEntry).put(0, -1900, flop=True)
    passive.strt(length=9928, xs='XS1').put(0, -1900)
   
    EdgeCouplerDoubleTaper(**parametersEdge).put()
    my_mmi21 = myMMI(posX = 0 , posY = -2300).put(0, -2300)
    my_mmi22 = myMMI(posX = 3000 , posY = -2100).put(3000, -2100)
    passive.sbend_p2p(pin1=my_mmi21.pin['b0'], pin2=my_mmi22.pin['a1']).put()
    finalConnect22 = passive.sbend().put(my_mmi22.pin['b0'])
    passive.strt(length=5383.358, xs='XS1').put(finalConnect22.pin['b0'])
    EdgeCouplerDoubleTaper(**parametersEdge).put()
    finalConnect21 = passive.sbend().put(my_mmi21.pin['b1'])
    passive.strt(length=8383.35, xs='XS1').put(finalConnect21.pin['b0'])
    EdgeCouplerDoubleTaper(**parametersEdge).put()
    EdgeCouplerDoubleTaper(**parametersEdgeEntry).put(0, -3000, flop=True)
    
    passive.strt(length=9928, xs='XS1').put(0, -3000)
   
    EdgeCouplerDoubleTaper(**parametersEdge).put()
    my_mmi31 = myMMI(posX = 0 , posY = -4000).put(0, -4000)
    my_mmi32 = myMMI(posX = 3000 , posY = -3800).put(3000, -3800)
    my_mmi33 = myMMI(posX = 6000 , posY = -3600).put(6000, -3600)
    passive.sbend_p2p(pin1=my_mmi32.pin['b0'], pin2=my_mmi33.pin['a1']).put()
    passive.sbend_p2p(pin1=my_mmi31.pin['b0'], pin2=my_mmi32.pin['a1']).put()
    
    
    finalConnect33 = passive.sbend().put(my_mmi33.pin['b0'])
    passive.strt(length=2383.35, xs='XS1').put(finalConnect33.pin['b0'])
    EdgeCouplerDoubleTaper(**parametersEdge).put()
    finalConnect32 = passive.sbend().put(my_mmi32.pin['b1'])
    passive.strt(length=5383.35, xs='XS1').put(finalConnect32.pin['b0'])
    EdgeCouplerDoubleTaper(**parametersEdge).put()
    finalConnect31 = passive.sbend().put(my_mmi31.pin['b1'])
    passive.strt(length=8383.355, xs='XS1').put(finalConnect31.pin['b0'])
    EdgeCouplerDoubleTaper(**parametersEdge).put()
    
    

    nd.export_gds()
main()
```
